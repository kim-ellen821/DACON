```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt 
import seaborn as sns    
```

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data = train.corr(), annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data = test.corr(), annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')
```

```python
plt.figure(figsize=(12,12))
for i in range(1,13):
    plt.subplot(3,4,i)
    sns.distplot(train.iloc[:,i])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,12))
for i in range(1,12):
    plt.subplot(3,4,i)
    sns.distplot(test.iloc[:,i])
plt.tight_layout()
plt.show()
```

```python
# train   quality    .
for i in range(11):
    fig = plt.figure(figsize = (12,6))
    sns.barplot(x = 'quality', y = train.columns[i+2], data = train)
```
